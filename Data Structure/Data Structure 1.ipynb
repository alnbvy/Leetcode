```python
# 566. Reshape the Matrix

# Approach 1: Using Queue
class Solution(object):
    def matrixReshape(self, mat, r, c):
        """
        :type mat: List[List[int]]
        :type r: int
        :type c: int
        :rtype: List[List[int]]
        """
        if (r * c != len(mat[0]) * len(mat)):
            return mat
        q = [element for row in mat for element in row]
        return [[q.pop(0) for _ in range(c)] for _ in range(r)]
# Time complexity : O(m⋅n) We traverse over m⋅n elements twice.

# Space complexity : O(m⋅n) The queue formed will be of size m⋅n.

# Tips: 1) learn the double for loop in one line. 2) learn the function .pop 
# 3) If a function returns something it will not proceed anymore.

# Approach 2: Without Using Extra Space
class Solution(object):
    def matrixReshape(self, mat, r, c):
        """
        :type mat: List[List[int]]
        :type r: int
        :type c: int
        :rtype: List[List[int]]
        """
        if (r * c != len(mat[0]) * len(mat)):
            return mat
        row = 0
        col = 0
        res = [[None for _ in range(c)] for _ in range(r)]
        for i in range(len(mat)):
            for j in range(len(mat[0])):
                res[row][col] = mat[i][j]
                col += 1
                if (col == c):
                    col = 0
                    row += 1
        return res
# Time complexity : O(m⋅n) We traverse over m⋅n elements twice.
# Space complexity : O(m⋅n) The queue formed will be of size m⋅n.

# Tips: 1) how to define an empty 2D list

# Approach 3: Using division and modulus
class Solution(object):
    def matrixReshape(self, mat, r, c):
        """
        :type mat: List[List[int]]
        :type r: int
        :type c: int
        :rtype: List[List[int]]
        """
        if (r * c != len(mat[0]) * len(mat)):
            return mat
        cnt = 0
        res = [[None for _ in range(c)] for _ in range(r)]
        for i in range(len(mat)):
            for j in range(len(mat[0])):
                res[cnt // c][cnt%c] = mat[i][j]
                cnt += 1
        return res
```

```python
# 118. Pascal's Triangle

class Solution(object):
    def generate(self, numRows):
        """
        :type numRows: int
        :rtype: List[List[int]]
        """
        res = []
        for r in range(numRows):
            row = [None for _ in range(r+1)]
            row[0], row[-1] = 1 , 1
            
            for i in range(1,r):
                row[i] = res[r-1][i-1] + res[r-1][i]
            res.append(row)

        return res
# Time complexity: O(numRows2)
# Space complexity: O(1)
# the input and output generally do not count towards the space complexity.
```
