```python
# 217. Contains Duplicate

class Solution(object):
    def containsDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        return False if (len(nums) == len(list(set(nums)))) else True
# Tip: note that True and False start with capital letter
```

```python
# 53. Maximum Subarray

class Solution(object):
    def maxSubArray(self, nums):
        mx = nums[0]
        running_sum = nums[0]
        for num in nums[1:]:
            running_sum = max(num,running_sum + num)
            mx = max(running_sum,mx)

        return mx
    
# You skipped the advanced approach of this problem
# Learn the space and time complexity of this problem
```
